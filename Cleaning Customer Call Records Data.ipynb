```python
import pandas as pd
```

# We will be cleaning customer call records data so that it's easier for those making calls. 


# Getting a sense of the data

```python
df = pd.read_excel("Customer Call List.xlsx")
```

```python
df.head(21)
```

# Drop duplicates first

```python
df = df.drop_duplicates()
```

```python
df
```

# Drop not_useful_column

```python
df = df.drop(columns = ["Not_Useful_Column"])

df
```

# Go column by column and see what needs to be cleaned/fixed

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...") #apply to only this column

df
```

```python
df["Last_Name"] = df["Last_Name"].str.rstrip("_") #apply to only this column
df["Last_Name"] = df["Last_Name"].str.lstrip("/") #apply to only this column

df
```

#alternatively we can do which will clean up 

df["Last_Name"].str.strip("123._/") #set of characters to remove as leading/trailing characters

# Standardising format (for phone numbers) 

```python
#1st we will use regex to remove all alphanumeric characters
#next we will standardise the phone numbers

df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True) #we will replace all non (^) alphanumeric characters with nothing so essentially removing them 

df
```

```python
#to standardise phone numbers - 1st convert to string 

df["Phone_Number"]= df["Phone_Number"].apply(lambda x: str(x))

df
```

```python
#then now standardise

df["Phone_Number"] = df["Phone_Number"].apply(lambda x:x[0:3]+'-'+x[3:6]+'-'+x[6:10])

df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')

df
```

# Splitting into columns for address

```python
df[["Street_Address","State","Zip_Code"]] =df["Address"].str.split(",",n=2,expand = True) #split at most 2 times so we get 3 outputs
#expand=True returns a dataframe instead of a series of lists
```

```python
df
```

# Standardising for Do_Not_Contact and Paying Customer

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')

df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')

df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
```

# Dealing with empty values

```python
df = df.fillna('')

df
```

```python
df["Address"] = df["Address"].str.replace("N/a",'')

df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("N/a",'')

df
```

# Drop those who said dont contact me

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"]=="Y":
        df.drop(x,inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"]=='':
        df.drop(x,inplace=True)

df
```

# Reset index (final dataset)

```python
df = df.reset_index(drop=True)

df
```
